```python
a = comb_am['predictions']*act_am_p
comb_res = []
for k, v in enumerate(comb_out):
  name = next(iter(comb_out[k].items()))[0]
  raw = next(iter(comb_out[k].items()))[1]['predictions']*next(iter(act_out[k].items()))[1]['predictions']
  trend = next(iter(mon_tre[k].items()))[1]['predictions']
  print(raw)
  profit = (trend[raw>0]*5000 - 5000).sum()
  comb_res.append((name,profit.item()))
comb_res
```

```python
a = comb_am['predictions']*act_am_p
mt_res = []
target=mt_out
for k, v in enumerate(target):
  name = next(iter(target[k].items()))[0]
  trend = next(iter(mon_tre[k].items()))[1]['predictions']
  raw = torch.rand(trend.shape)
  print(raw)
  profit = (trend[raw>0.5]*5000 - 5000).sum()
  print(profit)
  print(trend)
  mt_res.append((name,profit.item()))
mt_res
```

```python

mm_res = []
for k, v in enumerate(mm_out):
  name = next(iter(mm_out[k].items()))[0]
  raw = next(iter(mm_out[k].items()))[1]['predictions']*next(iter(act_out[k].items()))[1]['predictions']
  trend = next(iter(mon_tre[k].items()))[1]['predictions']
  print(raw)
  profit = (trend[raw>0]*5000 - 5000).sum()
  mm_res.append((name,profit.item()))
mm_res
```

```python
df=pd.DataFrame()
names = [i[0].replace('stocks_data/','').replace('.csv','') for i in mm_res]
df['names']=names
mm_val = [i[1] for i in mm_res]
comb_val = [i[1] for i in comb_res]
mt_val = [i[1] for i in mt_res]
df['Monte Carlo'] = mt_val
df['Momentum'] = mm_val
df['HiveMindInvestor'] = comb_val
df
```

# Comparison of overall score of all methods 

```python
n_groups = 6
bar_width = 0.1
alpha =0.8
fig, ax = plt.subplots(figsize=(15,10))
br1 = np.arange(n_groups)
br2 = [x + bar_width for x in br1]
br3 = [x + bar_width for x in br2]
br4 = [x + bar_width for x in br3]
br5 = [x + bar_width for x in br4]


companies = ['Amazon', 'Apple', 'Google', 'META', 'Netfilx', 'Tesla']

rects1 = plt.bar(br1, comb_os , bar_width,alpha=alpha,edgecolor = 'black',
color='r',
label='Combination')

rects2 = plt.bar(br2, st_os , bar_width,alpha=alpha,edgecolor = 'black',
color='g',
label='Sentiment')

rects3 = plt.bar(br3, ts_os, bar_width,alpha=alpha,edgecolor = 'black',
color='b',
label='Time-series')

rects4 = plt.bar(br4, mt_os , bar_width,alpha=alpha,edgecolor = 'black',
color='m',
label='Monte Carlo')

rects5 = plt.bar(br5, mm_os , bar_width,alpha=alpha,edgecolor = 'black',
color='y',
label='Momentum')

plt.xlabel('Stocks')
plt.ylabel('Overall Scores')
plt.title('Comparison of Scores of Methods by Stock')
plt.xticks([x + 2*bar_width for x in range(n_groups)], companies)
plt.legend()

plt.tight_layout()
plt.show()
```
