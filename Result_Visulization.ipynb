```python
import matplotlib.pyplot as plt
```

```python
time_windows = ['09/08-09/15','09/15-09/22','09/22-09/29','09/29-10/06','10/06-10/13','10/13-10/20','10/20-10/27','10/27-11/03','11/03-11/10','11/10-11/17','11/17-11/24']
amazon = result[0]['stocks_data/AMZN.csv']
amazon_score = amazon['overall_score']
amazon_prediction = amazon['predictions']
x_length = range(len(time_windows))
plt.figure(figsize=(15,10))
plt.plot(x_length, amazon_prediction,'o',label='Scores')
plt.plot(x_length,np.ones([len(time_windows),1]), label='One line')
plt.xticks(x_length,time_windows)
plt.xlabel('Time Windows')
plt.ylabel('Scores')
plt.legend(loc='best')
plt.title('Prediction vs. Time windows scores for Amazon')
```
